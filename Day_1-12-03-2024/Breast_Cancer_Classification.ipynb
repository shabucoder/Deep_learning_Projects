## BREAST CANCER CLASSIFICATION USING NEURAL NETWORK

### Importing the Required Dependencies


```python
import numpy as np
import pandas as pd
import sklearn.datasets
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

## Libraries required for building Neural Network
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

-> Data Collection and Processing

```python
## loading the data from sklearn
breast_cancer_dataset=sklearn.datasets.load_breast_cancer()
```

```python
print((breast_cancer_dataset))
```

```python
## converting data into the DataFrame
df=pd.DataFrame(breast_cancer_dataset.data, columns=breast_cancer_dataset.feature_names)
```

```python
## Printing the first 5 rows from the dataset
df.head(5)
```

```python
## adding the target column to the data frame

df['label']=breast_cancer_dataset.target
```

```python
## checking the shape of the dataset

df.shape
```

```python
## Getting some information about the data

df.info()
```

```python
df.groupby('label').mean()
```

```python
## divinding the dataset into dependent and independent features

X= df.drop('label', axis=1)
Y= df['label']
```

```python
## Now lets split above data into train and test data

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=.2,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape,Y.shape,Y_test.shape,Y_train.shape)
```

```python
Scaler=StandardScaler()
X_train_std=Scaler.fit_transform(X_train)
X_test_std=Scaler.transform(X_test)
```

```python
## Building Neural Network

## Setting up the layers of Neural Networks

model=keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),
    keras.layers.Dense(20, activation='relu'),
    keras.layers.Dense(2,activation='sigmoid')
])
```

```python
## compliling the Neural Nwtwork

model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
## training the model

history = model.fit(X_train_std,Y_train, validation_split=0.1, epochs=10)
```

```python
### Vizulizing the training Accuracy and validation accuracy


plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'lower right')
```

```python
### Vizulizing the training loss and validation loss

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'upper right')
```

```python
loss, accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
```
